# Data Quality Checks – Driver Performance Analysis

This notebook performs data quality validation on multiple tables used in the logistics profitability analysis.

Tables included:
- Fact tables: loads, trips, delivery_events
- Cost tables: fuel_purchases, maintenance_records, safety_incidents
- Dimension tables: drivers, trucks, trailers, customers, facilities, routes

The analysis focuses on the following tables that are relevant to the business case:

- Trips
- Loads
- Fuel_Purchases
- drivers
- delivery_events
- safety_incidents
- driver_monthly_metrics

## Import Libraries

```python
import pandas as pd
import numpy as np
from pathlib import Path 
```

## Load Data


```python
folder_path = Path(r"C:\Users\mory_\Downloads\Compressed\Logistic Database")

trips = pd.read_csv(folder_path/'trips.csv')
loads = pd.read_csv(folder_path / "loads.csv")
FUEL_PURCHASES = pd.read_csv(folder_path / "FUEL_PURCHASES.csv")
drivers = pd.read_csv(folder_path / "drivers.csv")
delivery_events = pd.read_csv(folder_path / "delivery_events.csv")
safety_incidents = pd.read_csv(folder_path / "safety_incidents.csv")
driver_monthly_metrics = pd.read_csv(folder_path / "driver_monthly_metrics.csv")
```

## Initial Data Inspection


```python
## Data Inspection and Data Quality Checks

def quality_check(df, name):
    
    print(f"\n---- {name} ----")
    
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
    
    print("\nMissing values:")
    print(df.isnull().sum())
    
    print("\nDuplicate rows:", df.duplicated().sum())

    print("\nChecking Data Types:")
    print(df.dtypes)

    print("\nChecking Invalid Values and basic Statistics:")
    print(df.describe())
```

```python
quality_check(trips, "Trips Table")
quality_check(loads, "Loads Table")
quality_check(drivers, "drivers Table")
quality_check(FUEL_PURCHASES, "FUEL_PURCHASES Table")
quality_check(delivery_events, "delivery_events Table")
quality_check(safety_incidents ,"safety incidents Table")
quality_check(driver_monthly_metrics, "driver_monthly_metrics Table")
```

## Data Cleaning


```python
trips = trips.dropna(subset=['driver_id'])
```

```python
FUEL_PURCHASES = FUEL_PURCHASES.dropna(subset=['driver_id'])
```

```python
safety_incidents = safety_incidents.dropna(subset=['driver_id'])
```

## Export Cleaned Data
Cleaned datasets were saved for further SQL analysis and Power BI dashboard development.

```python
trips.to_csv(folder_path/"trips_clean-2.csv", index=False)
FUEL_PURCHASES.to_csv(folder_path/'FUEL_PURCHASES_clean-2.csv',index=False)
safety_incidents.to_csv(folder_path/"safety_incidents_clean-2.csv", index=False)
```
